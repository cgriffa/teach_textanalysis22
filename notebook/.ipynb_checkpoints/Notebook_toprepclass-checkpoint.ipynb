## Prepare the pub dataset to use for students

```r
library(stringr)
df <- read.csv("../data/open_pubs_2020-07.csv", header = FALSE)
names(df) <- c("id", "name", "address", "postcode",   "easting", "northing", "latitude", "longitude", "local_authority")
head(df)
dim(df)
```

## Now make address dirty

```r
df$address_raw <- paste(df$address, df$postcode, sep = ", ")
df$address_raw <- str_replace_all(df$address_raw, ",", "")
df$latitude <- as.numeric(df$latitude)
df$longitude <- as.numeric(df$longitude)
df <- df[!is.na(df$latitude) & !is.na(df$longitude),]
```

```r
write.csv(df[,c("id", "name", "address_raw", "latitude", "longitude")], "../data/open_pubs_2020-07_forclass.csv")
```

## Prep speeches dataset


```r
library(tidyverse)
library(lubridate)
sp <- read.csv("../data/all_ECB_speeches.csv", sep ="|", quote = "")
head(sp)
```

```r
sp$contents <- str_replace_all(sp$contents, '"', '')
sp$id <- paste0("id_", str_pad(width = 4, pad = "0", string = row.names(sp)))
```

```r
sp$date2 <- ymd(sp$date)
dim(sp)
```

```r
out<- subset(sp, date2 - ymd(20111101) > 0)
head(out)
```

```r
names <- unique(out$speakers)
names <- names[names!= ""]
names <- str_replace_all(names, "[:punct:]", " ")
for (N in names) {
    print(N)
    n1 <- str_extract(N, "^[:alpha:]+[:space:]") 
    n2 <- str_extract(N, "[:space:][:alpha:]+$")
    out$contents <- str_replace_all(out$contents, "Speech by[:space:][:alpha:]+[:space:][:alpha:]+", "")
    out$contents <- str_replace_all(out$contents, str_trim(n1), "")
}
out$contents <- str_replace_all(str_lower(out$contents), "cœré", "")
```

```r
head(out)
write_delim(out, "../data/all_ECB_speeches20112022.csv", delim = "|")
```

## Prepare dictionary

```r
library(tidyverse)
library(tidytext)
library(SnowballC)
```

```r
rm(list = ls())
dict <- read.csv("../data/Loughran-McDonald_MasterDictionary_1993-2021.csv") %>% 
    subset(select = c(Word, Positive, Negative, Uncertainty, Litigious))
```

```r
head(dict)
```

```r
dict.bol <- (dict != 0)
dict <- data.frame( word = tolower(dict$Word), dict.bol[,2:5])
names(dict) <- c("word", "positive", "negative", "uncertainty", "litigious")
names(dict)
rm(dict.bol)
```

```r
head(dict)
```

```r
o <- which(as.vector(unlist(apply(dict[,2:5], 1, max))) > 0)
out <- dict[o,]
out[,2:5] %>% summarize_all(sum)

```

### Stemming the words


```r
#Stemming 
out.words <- out %>% mutate(stem = wordStem(word)) %>% 
    group_by(stem) %>% subset(select = - word) 
out.words[,-5] <- lapply(out.words[,-5], as.numeric)
out.words <- out.words %>% group_by(stem) %>% summarize_all("mean")
out.words[,-1] <- lapply(out.words[,-1], function(x){as.numeric(x > 0)})
```

```r
out.words[,-1] %>% summarize_all(c("min", "max")) 
dim(out.words)
```
